### M3 Tarefa Individual
Trabalho individual - análise de um artigo científico, de um podcast, ou e um vídeo selecionados pelos estudantes e validados pelo docente


#### Normas
- Cada estudante terá que indicar até **10/11/2022** qual o documento que pretende analisar. Para tal deverá ser preenchido o separador "tarefas" da nossa equipa no TEAMS. Apenas após a validação efetuada pelo professor, se poderá dar início à tarefa.
- O documento terá obrigatoriamente que estar relacionado com **desporto** e com **ciência de dados**, em qualquer das suas vertentes, como por exemplo a _Inteligência Artificial_ e o _Futebol_
- A entrega e apresentação desta tarefa terá lugar no dia **07/12/2022**
- O documento a entregar terá o **máximo de 500 palavras***
- A apresentação terá a **duração máxima de 6 minutos**

### M1 & M2 Trabalho de grupo

**Objetivo:** realização de uma Análise de Dados Exploratória, apartir de um documento .csv fornecido pelo professor

#### Normas
- Cada grupo terá que desenvolver código durante as aulas práticas e nas **horas de trabalho autónomo**, com supervisão e apoio do professor;
- A entrega e apresentação desta tarefa terá lugar no dia **12/01/2023**;
- O documento a entregar deverá ser um ficheiro de jupyter notebook, com todos os passos necessários para executar as análises propostas;
- Durante a apresentação, o grupo terá que iniciar o documento sem qualquer _output_, executando e explicando bloco-a-bloco;
- Todo o código deverá ser explicado em formato _Markdown_, sendo desejável o enquadramento das análises a realizar
- Para aferição da classificação M2-Autonomia, serão consideradas as participações ao longo das aulas práticas, assim como uma declaração dos restantes colegas de equipa.
